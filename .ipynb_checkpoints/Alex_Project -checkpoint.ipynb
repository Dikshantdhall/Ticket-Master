```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date

ticket_master = pd.read_csv("Event_Usa.csv")
ticket_master
```

```python
#Plot type of event from venue
Event_data = ticket_master["Event Category"].value_counts()
```

```python
Event_data.plot(kind="bar", facecolor="blue")

plt.title("Types of Events")
plt.xlabel("Event Category")
plt.ylabel("Number of Events")

plt.savefig("Types of Events")
```

```python
ticket_master.columns
```

```python
#Create a group based on promoter name
promoter_df = ticket_master.groupby('Promoter Name')

count_promoters = promoter_df['Promoter Name'].count()
count_promoters
```

```python
# Create a bar chart based off of the group series from before
count_chart = count_promoters.plot(kind='bar')

# Set the xlabel and ylabel using class methods
plt.title("Events Based on Promoters")
count_chart.set_xlabel("Promoters")
count_chart.set_ylabel("Number of Events")

plt.show()
plt.tight_layout()
#plt.savefig("Promoters List")
```

```python
#Minimum Pre Sale Ticket Price (for all of the events)
ticket_master.plot.scatter(x="Unnamed: 0", y="MinPrice", marker="o", facecolors="red", edgecolors="black", alpha=.9)
plt.title("Minimum Presale Cost ")
plt.xlabel("Each Event was Given a Numeric Value")
plt.ylabel("Minimum Price Range")


```
